# Import Data and Libraries

```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns
df = pd.read_csv('mtcars.csv')
df
#Model, miles per gallon, # of cylinders, displacement(cu.in.), horsepower, rear axle ratio, weight(1000s lb), 1/4 mile time
#V engine (0) or straight(1), automatic (0) or manual (1) transmission, # of forward gears, # of carburetors
```

### Pairplot

```python
#Simple! Just passing the dataframe will create a pairplot for all numeric columns
sns.pairplot(df)
```

```python
###
#What if we want to choose only certain columns?
#Simply slice the dataframe as desired
#If you want to make the plot more interesting, you can set the hue equal to some categorical column
#To show a breakdown based on that discrete data
###
sns.pairplot(df.iloc[:, 0:6], hue='cyl', palette="hls")
```

### Scatter Plot

```python
#Simply set the x and y parameters as the column variables you want to plot
sns.scatterplot(x=df['hp'], y=df['mpg'])
```

```python
#Hue may also be used here to communicate additional data
sns.scatterplot(x=df['hp'], y=df['mpg'], hue=df['am'])
```

```python
#The regplot method is similar to a scatter plot, but adds and plots a simple linear regression
sns.regplot(x=df['hp'], y=df['mpg'])
```

```python
#lmplot is similar to regplot, but allows for fitting regression models accross conditional subsets
sns.lmplot(x='hp', y='mpg', hue='vs', palette='Set1', data=df, ci=False, markers=["o", "x"])
```

# Pie Chart

```python
###
#Here is an example of creating a custom pie chart
#We define a list of some data and a list of labels, as well as our own color palette
#Then we simply call the plt.pie method and fill in the appropriate params
###
data = [35, 21, 29, 39, 11]
labels = ['slice 1', 'slice 2', 'slice 3', 'slice 4', 'slice 5']
colors = sns.color_palette('bright')
plt.pie(data, labels=labels, colors = colors, autopct = '%0.0f%%')
plt.show()
```

```python
#We can use the value_counts method to easily construct a pie chart based on some discrete column
df.cyl.value_counts().plot(kind='pie', autopct='%1.2f%%', shadow=True)
```

# Correlation Heatmap

```python
###
#Simply call the seaborn .heatmap method and pass the .corr() method on your dataframe as the parameter.
#The annot parameter includes the numbers on the graph (correlation coefficient)
###
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True)
```
